```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Ts = 0.0078125
Fs=1/Ts
```

```python
df = pd.read_excel('data00.xlsx', header=None, engine='openpyxl')

t = df[0]  # Tiempo (Columna A)
s1 = df[1] # Señal Original (Columna B)
s2 = df[2] # Señal Sobre-muestreada (Columna C)
s3 = df[3] # Señal Sub-muestreada (Columna D)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Gráfica 1
ax1.plot(t, s1, color='blue')
ax1.set_title('Señal Original')
ax1.set_ylabel('Amplitud')
ax1.grid(True)

# Gráfica 2
ax2.plot(t, s2, color='green')
ax2.set_title('Señal Sobre-muestreada')
ax2.set_ylabel('Amplitud')
ax2.grid(True)

# Gráfica 3
ax3.plot(t, s3, color='red')
ax3.set_title('Señal Sub-muestreada')
ax3.set_ylabel('Amplitud')
ax3.set_xlabel('Tiempo (s)')
ax3.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(t, s1, label='Señal Original', color='blue', alpha=0.7)
plt.plot(t, s2, label='Señal Sobre-muestreada', color='green', alpha=0.7)
plt.plot(t, s3, label='Señal Sub-muestreada', color='red', alpha=0.7)

plt.title('Comparación de las 3 Señales')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

plt.show()
```
